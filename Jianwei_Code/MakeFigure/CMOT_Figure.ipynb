```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# ---------------------
# 1. 生成示例数据
# ---------------------
# x 轴 (P/P_max)，对数分布
x = np.array([0.03, 0.1, 1])

# 左 y 轴数据：温度 T(mK)（示例用随机值）
T_data_1 = np.array([8, 19.4 , 232])

# 右 y 轴数据：云团尺寸 σ(mm)（示例用随机值）
sigma_data = np.array([0.43,0.49,1.01])

# 嵌入子图数据：激发态占比(%)（示例用随机值）
x_inset = np.array([0.1, 0.2, 0.5, 1.0])
excited_state_population = np.array([5, 10, 15, 20])  # 单位：%

# ---------------------
# 2. 创建主图
# ---------------------
fig, ax1 = plt.subplots(figsize=(6, 5))

# 设置 x 轴为对数坐标
ax1.set_xscale('log')
# 设置左 y 轴为对数坐标（温度）
ax1.set_yscale('log')

# 绘制温度数据 (蓝色圆点, 红色方块仅作示例)
ax1.plot(x, T_data_1, 'o', color='blue', label='T Data 1')

# ---------------------
# 3. 创建第二个 y 轴
# ---------------------
ax2 = ax1.twinx()  # 与 ax1 共享 x 轴
# 绘制 σ 数据 (红色方块)
ax2.plot(x, sigma_data, 's', color='red', label=r'$\sigma$ Data')

# ---------------------
# 4. 设置坐标轴标签、范围和标题
# ---------------------
ax1.set_xlabel(r'$P/P_{\max}$', fontsize=12)
ax1.set_ylabel(r'$T\ (\mathrm{mK})$', fontsize=12)
ax2.set_ylabel(r'$\sigma\ (\mathrm{mm})$', fontsize=12)

# 如果需要，可以设置坐标轴的范围：
#ax1.set_xlim(0.01, 10)         # x 轴范围 (对数坐标)
#ax1.set_ylim(0.5, 100)          # T 轴范围 (对数坐标示例)
#ax2.set_ylim(0, 2.2)            # σ 轴范围 (线性)

# 图标题
plt.title('Sn Compressed MOT Sequence', fontsize=14)

# ---------------------
# 5. 嵌入子图 (Inset)
# ---------------------
# inset_axes 可以方便地在主图上放置一个小窗
inset_ax = inset_axes(ax1, 
                      width="40%",   # 宽度：相对于 ax1 的百分比
                      height="30%",  # 高度：相对于 ax1 的百分比
                      loc='upper center')  # 位置可自行选择

# 在 inset 上绘制激发态占比
inset_ax.plot(x_inset, excited_state_population, 'o-', color='orange', label='Excited State %')
inset_ax.set_xscale('log')  # 若需要对数坐标
inset_ax.set_xlabel(r'$P/P_{\max}$', fontsize=8)
inset_ax.set_ylabel('Excited State Population (%)', fontsize=8)
inset_ax.tick_params(axis='both', which='major', labelsize=8)

# 视情况可设置 inset 坐标范围
inset_ax.set_xlim(0.08, 1.2)
inset_ax.set_ylim(0, 25)

# ---------------------
# 6. 图例与整体排版
# ---------------------
# 如果想在主图上显示多条曲线的图例，可以合并 ax1 和 ax2 的句柄
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='lower left')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import pandas as pd  # Import pandas for data handling

from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.gridspec import GridSpec


# 1. Data
# x-axis (P/P_max) in log scale
x = np.array([0.03, 0.1, 1])

# Temperature data T (mK)
T_data_1 = np.array([8, 19.4 , 232])

# Sigma data (mm)
sigma_data = np.array([0.43,0.49,1.91])

# Inset data: Excited state population (%)
excited_state_population = np.array([0.8,2.58,8.1])

# Read Trapped Trajectories in Red_MOT 
df_traj1 = pd.read_csv('SnMOTPaper/MOTTrajectoryYesScatter'+'20250405_0246'+'.csv', header=None,names=['velocity','position'])
df_traj2 = pd.read_csv('SnMOTPaper/MOTTrajectoryYesScatter'+'20250406_2311'+'.csv', header=None,names=['velocity','position'])
df_traj3 = pd.read_csv('SnMOTPaper/MOTTrajectoryYesScatter'+'20250407_0941'+'.csv', header=None,names=['velocity','position'])

# 2) Create Figure
fig = plt.figure(figsize=(10, 4))  # overall figure size in inches
# Create a main GridSpec with 3 rows and 2 columns.
# The top two rows hold panels (a)-(d), and the bottom row is reserved for panel (e)
gs = GridSpec(nrows=1, ncols=2, figure=fig, wspace=0.5)
 
# ------------------------------------------------------------------
# ============ SubFigure (a) ============
# ------------------------------------------------------------------
ax1 =fig.add_subplot(gs[0, 0])           
# Set x-axis to log scale
ax1.set_xscale('log')
# Set y-axis to log scale for temperature
ax1.set_yscale('log')

# Plot temperature data on ax1
ax1.plot(x, T_data_1, 'o', color='blue', label='T Data 1')


# Create a second y-axis (ax2) for sigma
ax2 = ax1.twinx()
ax2.plot(x, sigma_data, 's', color='red', label=r'$\sigma$ Data')

# Set axis labels and ranges, and assign colors to the y-axes
ax1.set_xlabel(r'$P/P_{\max}$', fontsize=8)
ax1.set_ylabel(r'$T\ (\mathrm{mK})$', fontsize=8, color='blue')
ax2.set_ylabel(r'$\sigma\ (\mathrm{mm})$', fontsize=8, color='red')

# Set y-axis tick colors to correspond with the data
ax1.tick_params(axis='y', labelcolor='blue')
ax2.tick_params(axis='y', labelcolor='red')

ax1.set_xlim(0.02, 1.2)   # x-axis range
ax1.set_ylim(5, 300)    # Temperature range (log scale)
ax2.set_ylim(0.3, 2.1)      # Sigma range (linear)
ax1.set_title("(a)", loc='left', fontsize=12, fontweight='bold')

#Set the main title
#plt.title('Sn Compressed MOT Sequence', fontsize=14)

# Create an inset plot on ax1
inset_ax = inset_axes(ax1, 
                      width="45%",   # Width relative to ax1
                      height="40%",  # Height relative to ax1
                      loc='upper center')  # Position of the inset

# Plot excited state population in the inset
inset_ax.plot(x_inset, excited_state_population, 'o-', color='orange', label='Excited State %')
inset_ax.set_xscale('log')
inset_ax.set_xlabel(r'$P/P_{\max}$', fontsize=8)
inset_ax.set_ylabel('Excited State Population (%)', fontsize=8,loc='center')
inset_ax.tick_params(axis='both', which='major', labelsize=8)
#inset_ax.set_xlim(0.08, 1.2)
#inset_ax.set_ylim(0, 25)

# ------------------------------------------------------------------
# ============ SubFigure (b) ============
# ------------------------------------------------------------------
ax_b = fig.add_subplot(gs[0, 1])
ax_b.plot( df_traj1['position'], df_traj1['velocity'],color='blue',linewidth=1)
ax_b.plot( df_traj2['position'], df_traj2['velocity'],color='red',linewidth=1)
ax_b.plot( df_traj3['position'], df_traj3['velocity'],color='green',linewidth=1)
ax_b.set_ylabel(r'Velocity ($m/s$)', fontsize=9)
ax_b.set_xlabel(r'Position($mm$)', fontsize=9)
ax_b.tick_params(labelsize=8)
ax_b.set_title("(b)", loc='left', fontsize=12, fontweight='bold')
#ax_e.grid(True, linestyle='--', alpha=0.5)

# ===================== 3. Finalize Layout and Save Figure =====================
# Save the final figure as a PDF with a tight bounding box and a transparent background.
plt.savefig("PaperFig/SnMOT_CMOT.pdf", format='pdf', transparent=True)
#  Final adjustments and display
plt.tight_layout()
plt.show()
```
