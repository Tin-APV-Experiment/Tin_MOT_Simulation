### Blue MOT Heatmap

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from matplotlib.colors import LinearSegmentedColormap
from mpl_toolkits.axes_grid1 import make_axes_locatable  # for consistent colorbar height

# ====== 1. Read CSV data ======
# Assumes: first row = x coordinates, first column = y coordinates, top-left cell = NaN
df_raw = pd.read_csv('SnMOTPaper/Heatmap20250310_0450.csv', header=None)

# Use the first row as column headers, and the first column as row index
df_raw.columns = df_raw.iloc[0]
df_raw.index = df_raw.iloc[:, 0]

# Remove the NaN row and column
df = df_raw.drop(index=df_raw.index[0], columns=df_raw.columns[0])

# Convert axes labels to float (physical coordinates)
df.columns = df.columns.astype(float)  # x-axis labels
df.index = df.index.astype(float)      # y-axis labels

# ====== 2. Extract data matrix & coordinate arrays ======
data = df.to_numpy()                  # 2D data matrix
x_coords = df.columns.to_numpy()     # x-coordinates
y_coords = df.index.to_numpy()       # y-coordinates

# ====== 3. Create heatmap using imshow ======
fig, ax = plt.subplots(figsize=(8, 6))

# Create a custom red-white-blue colormap
red_blue = LinearSegmentedColormap.from_list("red_blue", ["red", "white", "blue"])
cmap_custom =LinearSegmentedColormap.from_list(
    "cmap_custom",
    [
        (0.0,  "#00008B"),  # darkblue
        (0.25, "#0000FF"),  # blue
        (0.50, '#D3D3D3'),  # gray
        (0.75, "#FF0000"),  # red
        (1.0,  "#8B0000"), # darkred
    ]
)

# Use axes divider to match colorbar height with main plot
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)  # Add colorbar axis to the right

# Plot the heatmap
im = ax.imshow(data,
               cmap=cmap_custom,
               origin = "lower",
               extent=[x_coords[0], x_coords[-1], y_coords[0], y_coords[-1]],
               aspect="auto")

# ====== 4. Add colorbar ======
cbar = fig.colorbar(im, cax=cax)
cbar.set_label(r"Acceleration $a(r,v)$ / m/s$^2$")

# ====== 5. Axis labels ======
ax.set_xlabel(r"Position / $mm$")
ax.set_ylabel(r"Velocity / $m/s$")

# ====== 6. Fixed tick intervals ======
ax.xaxis.set_major_locator(MultipleLocator(0.5))
ax.yaxis.set_major_locator(MultipleLocator(0.4))

plt.tight_layout()
plt.show()
```

Acceleration Vs Position

Another Template with Major and Minor Grids

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, AutoMinorLocator

# === 1. Read the CSV file ===
df_ar = pd.read_csv('SnMOTPaper/AccelVsPos20250310_0450.csv', header=None)
df_ar.columns = ['Position', 'Acceleration']

x = df_ar['Position']
y = df_ar['Acceleration']

# === 2. Create a figure and axis ===
fig, ax = plt.subplots(figsize=(5, 5))

# === 3. Plot the data ===
ax.plot(x, y, linestyle='-', color='red', label='a(x)')

# === 4. Customize the axis ticks ===
# Major ticks every 0.5 on the x-axis (adjust as needed)
ax.xaxis.set_major_locator(MultipleLocator(0.5))
# Minor ticks: automatically determined subdivisions
ax.xaxis.set_minor_locator(AutoMinorLocator())

# Similarly for the y-axis (customize to your data range)
ax.yaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(AutoMinorLocator())

# === 5. Enable and style the grid ===
# Make sure the grid is behind the data
ax.set_axisbelow(True)

# Major grid lines: solid, thinner black
ax.grid(which='major', color='black', linestyle='-', linewidth=0.5, alpha=0.7)

# Minor grid lines: dashed, lighter gray
ax.grid(which='minor', color='gray', linestyle='--', linewidth=0.5, alpha=0.5)

# === 6. Labeling and layout ===
ax.set_xlabel('Position (mm)')
ax.set_ylabel('Acceleration (m/s²)')
ax.legend()
plt.tight_layout()
plt.show()
```

### Paper-Figure For Blue MOT Simulation
Four subfigures:
1. (a)Heatmap
2. (b)Acceleration vs Position
3. (c) Many-Particle Simulation: Two stable points in free space
4. (d)Acceleration vs Velocity

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  # Import pandas for data handling
from matplotlib.colors import LinearSegmentedColormap, TwoSlopeNorm
from matplotlib.ticker import MultipleLocator, ScalarFormatter
from mpl_toolkits.axes_grid1 import make_axes_locatable

# ===================== 1. Read Data =====================
# Read acceleration vs. position data (no header in file)
df_ar = pd.read_csv('SnMOTPaper/AccelVsPos20250310_0450.csv', header=None)
df_ar.columns = ['Position', 'Acceleration']  # Define column names for clarity

# Read acceleration vs. velocity data (no header in file)
df_av = pd.read_csv('SnMOTPaper/AccelVsVel20250310_0450.csv', header=None)
df_av.columns = ['Velocity', 'Acceleration']

# Read final positions and velocities along with additional parameters (header is present)
df_final = pd.read_csv('SnMOTPaper/MOTTempAndSize_ManyParticle_20250310_0450.csv', header=0)
df_final.columns = ['FinalPosition', 'FinalVelocity', 'a' , 'b']

# ====== Read CSV data for Heatmap======
# Assumes: first row = x coordinates, first column = y coordinates, top-left cell = NaN
df_raw = pd.read_csv('SnMOTPaper/Heatmap20250310_0450.csv', header=None)
# Use the first row as column headers, and the first column as row index
df_raw.columns = df_raw.iloc[0]
df_raw.index = df_raw.iloc[:, 0]
# Remove the NaN row and column
df = df_raw.drop(index=df_raw.index[0], columns=df_raw.columns[0])
# Convert axes labels to float (physical coordinates)
df.columns = df.columns.astype(float)  # x-axis labels
df.index = df.index.astype(float)      # y-axis labels
# Extract data matrix & coordinate arrays 
data = df.to_numpy()                  # 2D data matrix
x_coords = df.columns.to_numpy()     # x-coordinates
y_coords = df.index.to_numpy()       # y-coordinates

# ===================== 2. Create Figure and Subplots =====================
# Create the overall figure with a specified size (width x height in inches)
fig = plt.figure(figsize=(7, 6))

# Define custom positions for each subplot as [left, bottom, width, height] in relative coordinates (0 to 1)
# These positions manually arrange 4 subplots on the figure.
positions = [
    [0.05, 0.59, 0.36, 0.36],  # Subplot (a)
    [0.59, 0.59, 0.36, 0.36],  # Subplot (b)
    [0.05, 0.05, 0.36, 0.36],  # Subplot (c)
    [0.59, 0.05, 0.36, 0.36]   # Subplot (d)
]

# Labels for each subplot (not used for plotting directly but can be useful for further customization)
labels = ['(a)', '(b)', '(c)', '(d)']

# ===================== 3. Plot Subplot (a): Heatmap =====================
# Note: The variables 'data', 'x_coords', and 'y_coords' are assumed to be defined elsewhere.
ax_a = fig.add_axes(positions[0])

# Create a custom colormap that transitions from blue to white to red
red_blue = LinearSegmentedColormap.from_list("red_blue", ["blue", "white", "red"])
cmap_custom =LinearSegmentedColormap.from_list(
    "cmap_custom",
    [
        (0.0,  "#00008B"),  # darkblue
        (0.25, "#0000FF"),  # blue
        (0.50, "#FFFFFF"),  # white
        (0.75, "#FF0000"),  # red
        (1.0,  "#8B0000"), # darkred
    ]
)
cmap_custom_r = cmap_custom.reversed()

# Normalize the colormap with a center value of 0 so that the midpoint of the colormap corresponds to 0.
norm = TwoSlopeNorm(vmin=np.min(data), vcenter=0, vmax=np.max(data))

# Create an axes divider for placing the colorbar next to the heatmap
divider = make_axes_locatable(ax_a)
cax = divider.append_axes("right", size="5%", pad=0.05)

# Display the heatmap using imshow. 'extent' sets the coordinate system using x_coords and y_coords.
im = ax_a.imshow(data, cmap='coolwarm', norm=norm, origin="lower",
                 extent=[x_coords[0], x_coords[-1], y_coords[0], y_coords[-1]],
                 aspect="auto")

# Add a colorbar to the heatmap and label it
cbar = fig.colorbar(im, cax=cax)
cbar.set_label(r"Acceleration ($m/ms^2$)", fontsize=9)
cbar.ax.tick_params(labelsize=8)

# Set axis labels and tick parameters for the heatmap subplot
ax_a.set_xlabel(r"Position ($mm$)", fontsize=9)
ax_a.set_ylabel(r"Velocity ($m/s$)", fontsize=9)
ax_a.tick_params(labelsize=8)
ax_a.text(-0.15, 1.08, '(a)', transform=ax_a.transAxes, fontsize=12, fontweight='bold')

# ===================== 4. Plot Subplot (b): Acceleration vs. Position =====================
ax_b = fig.add_axes(positions[1])  # Create subplot (b) at the defined position
# Plot acceleration vs. position from the first dataset
ax_b.plot(df_ar['Position'], df_ar['Acceleration'], color='red', linestyle='-', linewidth=1)
ax_b.set_xlabel(r'Position ($mm$)', fontsize=9)
ax_b.set_ylabel(r'Acceleration ($m/ms^2$)', fontsize=9)
ax_b.tick_params(labelsize=8)
#ax_b.grid(True, alpha=0.3)  # Add a grid for easier reading
ax_b.text(-0.15, 1.1, '(b)', transform=ax_b.transAxes, fontsize=12, fontweight='bold')

# ===================== 5. Plot Subplot (c): Trajectory Plot =====================
ax_c = fig.add_axes(positions[2])  # Create subplot (c) at the defined position
# Scatter plot of final positions vs. final velocities from the third dataset
ax_c.scatter(df_final['FinalPosition'], df_final['FinalVelocity'], s=10, color='blue', alpha=0.6)
ax_c.set_xlabel(r'Final Position Center ($mm$)', fontsize=9)
ax_c.set_ylabel(r'Final Square-Root Velocity ($m/s$)', fontsize=9)

# Customize the x-axis: set major ticks every 0.5 units
ax_c.xaxis.set_major_locator(MultipleLocator(0.5))
# Customize the y-axis: set major ticks every 0.0005 units
ax_c.yaxis.set_major_locator(MultipleLocator(0.0005))
# Format the y-axis tick labels in scientific notation for clarity
ax_c.ticklabel_format(axis='y', style='sci', scilimits=(-2, 2))
ax_c.yaxis.offsetText.set_fontsize(8)
# Optionally, format the x-axis tick labels as plain numbers
ax_c.ticklabel_format(axis='x', style='plain')
ax_c.tick_params(labelsize=8)
#ax_c.grid(True, alpha=0.3)  # Add a light grid
ax_c.text(-0.15, 1.08, '(c)', transform=ax_c.transAxes, fontsize=12, fontweight='bold')

# ===================== 6. Plot Subplot (d): Velocity vs. Acceleration =====================
ax_d = fig.add_axes(positions[3])  # Create subplot (d) at the defined position
# Plot velocity vs. acceleration from the second dataset
ax_d.plot(df_av['Velocity'], df_av['Acceleration'], color='red', linestyle='-', linewidth=1)
ax_d.set_xlabel(r'Velocity ($m/s$)', fontsize=9)
ax_d.set_ylabel(r'Acceleration  ($m/ms^2$)', fontsize=9)
ax_d.tick_params(labelsize=8)
#ax_d.grid(True, alpha=0.3)  # Add a grid for easier reading
ax_d.text(-0.15, 1.1, '(d)', transform=ax_d.transAxes, fontsize=12, fontweight='bold')
# ===================== 7. Finalize Layout and Save Figure =====================
# Save the final figure as a PDF with a tight bounding box and a transparent background.
plt.savefig("PaperFig/SnMOT_BlueMOT.pdf", format='pdf', bbox_inches='tight', transparent=True)
plt.show()

```
